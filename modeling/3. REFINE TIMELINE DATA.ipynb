# Check types

```python
kinds = "Additional"
tier = "MASTER"

# Additional, Train
#df = pd.read_csv("./Timeline DataFrames/{0}/{1}/Train/{1}{2}.csv".format(kinds, tier, 15))

# Additional, Test
df = pd.read_csv("./Timeline DataFrames/{0}/{1}/Test/{1}{2}.csv".format(kinds, tier, 15))

# Original
#df = pd.read_csv("./Timeline DataFrames/{0}/{1}/{1}{2}.csv".format(kinds, tier, 15))

print(df['blueFirstTowerLane'])
print("Before delete:")
print("index, col name, type")
for idx, col in enumerate(df.columns):
    print("{} {}: {}".format(idx, col, type(df[col].iloc[2])))
del df["gameId"]
del df["blueWins"]
del df["redWins"]
for col in df.columns:
    if str(type(df[col].iloc[2])) == "<class 'str'>":
        del df[col]
print("\nindex, col name, type")
for idx, col in enumerate(df.columns):
    print("{} {}: {}".format(idx, col, type(df[col].iloc[2])))
```

```python
start, end = 1, 50
for tl in range(start, end+1):
    print("-------------------------------------------------------------")
    print(" >> Processing : {} {} {} minute".format(kinds, tier, tl))
    
    # Additional, Train
    #df = pd.read_csv("./Timeline DataFrames/{0}/{1}/Train/{1}{2}.csv".format(kinds, tier, tl))

    # Additional, Test
    df = pd.read_csv("./Timeline DataFrames/{0}/{1}/Test/{1}{2}.csv".format(kinds, tier, tl))

    # Original
    #df = pd.read_csv("./Timeline DataFrames/{0}/{1}/{1}{2}.csv".format(kinds, tier, tl))
    
    timeline_df = pd.DataFrame(columns=diff_columns)
    timeline_df["game_id"] = df["gameId"]
    timeline_df["blue_win"] = df["blueWins"]
    del df["gameId"]
    del df["blueWins"]
    del df["redWins"]
    for col in df.columns:
        if str(type(df[col].iloc[2])) == "<class 'str'>":
            del df[col]
    
    raw_columns = list(df.columns)
    offset = len(raw_columns)//2 # 48//2 = 24
    
    for idx in range(offset): # 0 to 23, offset=24
        bf, rf = raw_columns[idx], raw_columns[idx+offset]
        lst = df[bf] - df[rf]
        timeline_df[diff_columns[idx+2]] = lst
        print("{} - {} → {} done".format(bf, rf, diff_columns[idx+2]))
    
    # Additional, Train
    #file_name = "./Refined Timeline DataFrames/{0}/{1}/Train/{1}{2}.csv"
    
    # Additional, Test
    file_name = "./Refined Timeline DataFrames/{0}/{1}/Test/{1}{2}.csv"
    
    # Original
    #file_name = "./Refined Timeline DataFrames/{0}/{1}/{1}{2}.csv"
    
    timeline_df.to_csv(file_name.format(kinds, tier, tl), index=False)
    print("shape: {} saved".format(timeline_df.shape))
    print("-------------------------------------------------------------")
print("\nAll works have been finished")
```
