## Make dataset(not split train/dataset)

```python
numMatches = len(endTimes)
print(" > {} {} real rnn dataset processing start".format(kinds, tier))
print("Whole game size is:", numMatches)
import time
#RNN_m 모델을 학습시키기 위한 데이터, 타겟 생성
print("-----------------------------------------------------")

"""
# Additional, Train
file_name_data = "RNN Real Dataset/{0}/{1}/Train/Data/{1}{2}.pkl"
file_name_target = "RNN Real Dataset/{0}/{1}/Train/Target/{1}{2}.pkl"
"""

#"""
# Additional, Test
file_name_data = "RNN Real Dataset/{0}/{1}/Test/Data/{1}{2}.pkl"
file_name_target = "RNN Real Dataset/{0}/{1}/Test/Target/{1}{2}.pkl"
#"""

"""
# Original
file_name_data = "RNN Real Dataset/{0}/{1}/Data/{1}{2}.pkl"
file_name_target = "RNN Real Dataset/{0}/{1}/Target/{1}{2}.pkl"
"""

total_time = time.time()
for m in range(2, 51):
    start_time = time.time()
    print(" >> Processing: RNN_{} model's data & target".format(m))
    dataset, eachTarget = [], []
    for idx in range(numMatches):
        endTime = endTimes[idx]
        miniset = []
        if m <= endTime:
            for k in range(1, m+1):
                miniset.append(eval("df_{}".format(k)).iloc[idx])
            dataset.append(np.array(miniset))
            eachTarget.append(target[idx])
    dataset = np.array(dataset)
    with open(file_name_data.format(kinds, tier, m), "wb") as f:
        pickle.dump(dataset, f)
    with open(file_name_target.format(kinds, tier, m), "wb") as f:
        pickle.dump(eachTarget, f)
    print("Saved. dataset.shape: {}, len(target): {}".format(dataset.shape, len(eachTarget)))
    print("Wait time: {} second(s)".format(round(time.time()-start_time, 2)))
    print("-----------------------------------------------------")
print("\nTotal wait time: {} second(s)".format(round(time.time()-total_time, 2)))
print("All of works have been finished")
```
