### Process Original 집계데이터

```python
kinds = 'Original'
divisions = ['III', 'II', 'I']
file = 6
for division in divisions:
    total_time = time.time()
    df = pd.DataFrame()
    IDWinLoseList = pd.read_pickle("../IDWinLose List/{0}/{1}/{1}{2}IDWinLoseList".format(kinds, tier, division))
    print("{0}/{1}/{1}{2}IDWinLoseList loaded".format(kinds, tier, division))
    print("-----------------------------------------------------------")
    listLength = len(IDWinLoseList)
    f_start_time = time.time()
    for idx, IDWinLose in enumerate(IDWinLoseList):
        game_id, bw, rw, data = IDWinLose
        if idx % 1000 == 0:
            print(" >> loop location: {}/{} / processing: {} %".format(idx, listLength, round(idx*100/listLength, 1)))
            try: print("Wait time: {} second(s)".format(round(time.time()-start_time, 1)))
            except: pass
            finally: start_time = time.time()
        for i in range(10):
            try:
                pre_df = pd.json_normalize(data['participants'][i])
                refined_df = pd.DataFrame()
                for col in columns:
                    refined_df[col] = pre_df[col]
                df = pd.concat([df, refined_df])
                #print("len(columns): {} / df.shape: {}".format(len(columns), df.shape))
            except:
                pass
    print("-----------------------------------------------------------")
    print("Final dataframe.shape: {}".format(df.shape))
    df.to_csv("Aggregated DataFrames/{0}/{0}{1}.csv".format(tier, file))
    file += 1
    print("{0}/{1}IDWinLoseList{2} finished\n".format(kinds, tier, division))
print("Total wait time: {} second(s)".format(round(time.time()-total_time, 1)))
print("All works have been finished.")
```
